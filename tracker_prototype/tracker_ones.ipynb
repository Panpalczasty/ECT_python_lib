```python
import tracker as tr

radius = 200
offset = radius/20
ect_offset = radius/10

loader = tr.FilepathLoader(
    radius=radius,
    offset=offset
)

matcher = tr.BasicMatcher(
    bp_thresh = 0.2
)

tracker = tr.BasicTracker(
    template_path="../tracker_tests/1_template.png",
    radius = radius,
    offset = offset,
    ect_offset = ect_offset,
    matcher = matcher
)

tracker.setup()
```

### Scaling with center in fovea

```python
fpaths = [
    "../tracker_tests/1_scaled_50.png",
    "../tracker_tests/1_scaled_60.png",
    "../tracker_tests/1_scaled_70.png",
    "../tracker_tests/1_scaled_80.png",
    "../tracker_tests/1_scaled_90.png",
    "../tracker_tests/1_template.png",
    "../tracker_tests/1_scaled_110.png",
    "../tracker_tests/1_scaled_120.png",
]

for fpath in fpaths:

    image = loader.load(fpath)
    tracker.match(image)
    tracker.show_result(fpath.split("/")[-1])
```

### Rotating with center in fovea

```python
fpaths = [
    "../tracker_tests/1_template.png",
    "../tracker_tests/1_rotated_30.png",
    "../tracker_tests/1_rotated_60.png",
    "../tracker_tests/1_rotated_90.png",
    "../tracker_tests/1_rotated_120.png",
    "../tracker_tests/1_rotated_150.png",
    "../tracker_tests/1_rotated_180.png",
]

for fpath in fpaths:

    image = loader.load(fpath)
    tracker.match(image)
    tracker.show_result(fpath.split("/")[-1])
```
