```python
import tracker as tr
import ect
import matplotlib.pyplot as plt
import cv2
import numpy as np

radius = 200
offset = 10
ect_offset = 0

loader = tr.SequenceLoader(
    dirpath = "../sequence_test",
    root_name = "1_seq_",
    ftype = "png",
    index = 0
)

transformer = tr.UnfilteredTransformation(
    img_offset = offset,
    ect_offset = ect_offset
)

matcher = tr.RotShiftMatcher(
    transformer = transformer,
    center_fnc = tr.center_of_mass,
    bp_thresh = 0.1
)

tracker = tr.TrajectoryLPTracker(
    template_path = "../tracker_tests/1_template.png",
    matcher = matcher,
    radius = radius,
    offset = offset,
    ect_offset = ect_offset
)

tracker.setup()
```

```python
for image in loader.load_batch():
    tracker.match(image)

tracker.show_result("Detected Log-Polar Trajectory")
```
