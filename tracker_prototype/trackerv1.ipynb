```python
import tracker as tr
import ect
import matplotlib.pyplot as plt
import numpy as np

loader = tr.FilepathLoader()
transformer = tr.FilteredTransformation(
    img_offset = 20,
    ect_offset = 40,
    radius = 200
)

matcher = tr.CorrelationMatcher(
    transformer = transformer
)

tracker = tr.BasicTracker(
    template_path="../sample_imgs/black_e.png",
    matcher = matcher,
    transformer = transformer
)
tracker.setup()
```

```python
e_filepaths = [
    "../sample_imgs/black_e_small.png",
    "../sample_imgs/black_e_4.png",
    "../sample_imgs/black_e.png",
    "../sample_imgs/black_e_3.png",
    "../sample_imgs/black_e_big.png",
    "../sample_imgs/black_e_90deg.png",
    "../sample_imgs/white.png",
    "../sample_imgs/whitenoise.png",
    "../sample_imgs/e_template.png"
]

for e in e_filepaths:

    e_img = loader.load(e)

    tracker.match(e_img)
    tracker.show_result(e.split("/")[-1])
```
